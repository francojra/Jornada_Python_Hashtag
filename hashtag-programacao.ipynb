```python
# Importar base de dados
import pandas
tabela = pandas.read_csv("cancelamentos.csv")
```

```python
# Visualizar base de dados
display(tabela)
```

```python
display(tabela.info()) # Mostra se a tabela tem valores vazios
```

```python
# Tratamento de erros
# Tratar valores vazios
tabela = tabela.dropna() # Exclui células vazias
```

```python
display(tabela.info()) # Mostra mudanças na base 
```

```python
# Informação que não ajuda, te atrapalha. Devemos excluir colunas inúteis para as análises
# O comando drop exclui tanto linhas quanto colunas
tabela = tabela.drop("CustomerID", axis = 1) # axis = 1 indica exclusão de coluna e axis = 0 exclusão de linha
display(tabela)
```

```python
# Análise inicial da base de dados
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize = True))
```

```python
display(tabela["cancelou"].value_counts(normalize = True).map("{:.1%}".format)) # Mostra as porcentagens
```

```python
display(tabela.groupby("duracao_contrato").mean()) # Calcula a média para todas as colunas em relação a duração de contrato
```

```python
# Contrato mensal sempre cancela
```

```python
display(tabela.groupby("assinatura").mean())
```

```python
tabela = tabela[tabela["duracao_contrato"] != "Monthly"] # Retira todas linhas de contrato mensal
```

```python
# Analisando apenas os dois gráficos indicados, nós podemos manter na tabela as ligações ao call center que são menores 
# do que 5.

# E podemos manter também as informações onde os dias de atraso são menores do que 0.

# Com isso podemos visualizar a base de dados e fazer o cálculo novamente para verificar o percentual de cancelamento.

# Veja que só removendo as informações dessas duas colunas passamos de um percentual de quase 50% de cancelamento
# para 18,4%.
```

```python
tabela = tabela[tabela["ligacoes_callcenter"] < 5]
tabela = tabela[tabela["dias_atraso"] <= 20]
display(tabela)
```
